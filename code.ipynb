### Handling Categorical Data

Probability Ratio Encoding

```python
# Probability Ratio Encoding

Categ = ['NAME','REPORT_TYPE']
Final_trn
for i in Categ:
    dict_map = Final_trn[i].value_counts().to_dict()
    Final_trn[i+"_freq"] = Final_trn[i].map(dict_map)
    
    prob_dt = Final_trn.groupby([i])[i+"_freq"].mean()
    
    a = prob_dt.sum()
    prob_list = []
    for j in range(len(prob_dt)):
        b = prob_dt[j]
        d = b/a*100
        prob_list.append(d)
    
    prob_df = pd.DataFrame(prob_dt)
    prob_df[i+"_prob"]= prob_list
    Name_encoded = prob_df[i+"_prob"].to_dict()
    Final_trn[i+"_ENCODED"] = Final_trn[i].map(Name_encoded)
    Final_trn = Final_trn.drop(i+'_freq',axis=1)

drop_colm = []

for i in Final_trn.columns:
    if Final_trn[i].dtype == 'object':
        drop_colm.append(i)
        
Final_trn = Final_trn.drop(drop_colm,axis =1)
```

### Replacing NaN values with Mean, Median and Mode

for each columns having numerical values replacing the NaN values in them with Mean, Median and Mode.

```python
Final_trn.isnull().mean()
```

Ways to replace NaN values:

1.Mean                          with additional NaN_feature 

2.Median                        with additional NaN_feature

3.Mode                          with additional NaN_feature

4.Best of Mean Median Mode      with additional NaN_feature

5.Random Sampling               with additional NaN_feature

```python
#NaN Imputing ways

Final_trn_all = Final_trn.copy()
Final_trn_mean = Final_trn.copy()
Final_trn_median = Final_trn.copy()
Final_trn_mode = Final_trn.copy()
Final_trn_random = Final_trn.copy()

class impute:
    def impute_nan_all(self):
        
        
        median=Final_trn[self].median()
        mean = Final_trn[self].mean()
        mode = Final_trn[self].mode()[0]
    
        t = Final_trn[self]
    
        Final_trn_all[self+"_NAN"] = np.where(Final_trn[self].isnull(),1,0)

        t[self+"_mean"]=Final_trn[self].fillna(mean)
        t[self+"_median"]=Final_trn[self].fillna(median)
        t[self+"_mode"]=Final_trn[self].fillna(mode)
    
        std = Final_trn[self].std()
        std_mean = t[self+'_mean'].std()
        std_median = t[self+'_median'].std()
        std_mode = t[self+'_mode'].std()

        dicttt = {'mean' : abs(std - std_mean),
                  'median' : abs(std - std_median),
                  'mode' : abs(std - std_mode)}
        d = min(dicttt.values())
        metho = [k for k, v in dicttt.items() if v == d]
        Final_trn_all[self+'_'+metho[0]] = Final_trn[self].fillna(dicttt[metho[0]])
    

    def impute_nan_mean(self):
        mean = Final_trn[self].mean()
    
        Final_trn_mean[self+"_NAN"]=np.where(Final_trn[self].isnull(),1,0)
        Final_trn_mean[self+"_mean"]=Final_trn[self].fillna(mean)

        
    def impute_nan_median(self):
        median=Final_trn[self].median()
    
        Final_trn_median[self+"_NAN"]=np.where(Final_trn[self].isnull(),1,0)
        Final_trn_median[self+"_median"]=Final_trn[self].fillna(median)

        
    def impute_nan_mode(self):
        mode = Final_trn[self].mode()[0]
    
        Final_trn_mode[self+"_NAN"]=np.where(Final_trn[self].isnull(),1,0)
        Final_trn_mode[self+"_mode"]=Final_trn[self].fillna(mode)


#Random Value Imputation
    def impute_nan_random(self):
        Final_trn_random[self+"_random"] = Final_trn_random[self]
        Final_trn_random[self+"_NAN"]=np.where(Final_trn[self].isnull(),1,0)

        random_sample= Final_trn_random[self].dropna().sample(Final_trn_random[self].isnull().sum(),random_state=0)
        random_sample.index=Final_trn_random[Final_trn_random[self].isnull()].index
        Final_trn_random.loc[Final_trn_random[self].isnull(),self+"_random"]=random_sample
        
```

```python
# Dropping NaN values from these columns for final dataframe

def drop_nan_value_columns(i):
        
        drop_colm = []
        p = i.isnull().sum(axis = 0)
        for j in range(len(p)):
            if p[j] != 0:
                drop_colm.append(i.columns[j])       
        
        i = i.drop(drop_colm, axis =1)
        return i
```

```python
C_Names = ['HourlyDewPointTemperature','HourlyDryBulbTemperature','HourlyRelativeHumidity','HourlySeaLevelPressure','HourlyWindDirection','HourlyWindSpeed']
```

```python
New_Columns = {}
for j in C_Names:
    p = final_tst[j].values
    New_Columns[j] = []
    for i in p:
        try:
            a = float(i)
            New_Columns[j].append(a)
        except:
            i = i[:-1]
            a = float(i)
            New_Columns[j].append(a)
```

```python
c = final_tst.columns
convert_dict = {}
for i in c:
    if i != 'REPORT_TYPE' and i != 'NAME':
        convert_dict[i] = float
```

```python
convert_dict
```

```python
Final_tst = final_tst.astype(convert_dict)
```

```python
Final_tst.dtypes
```

```python
Categ = ['NAME','REPORT_TYPE']
Final_tst
for i in Categ:
    dict_map = Final_tst[i].value_counts().to_dict()
    Final_tst[i+"_freq"] = Final_tst[i].map(dict_map)
    
    prob_dt = Final_tst.groupby([i])[i+"_freq"].mean()
    
    a = prob_dt.sum()
    prob_list = []
    for j in range(len(prob_dt)):
        b = prob_dt[j]
        d = b/a*100
        prob_list.append(d)
    
    prob_df = pd.DataFrame(prob_dt)
    prob_df[i+"_prob"]= prob_list
    Name_encoded = prob_df[i+"_prob"].to_dict()
    Final_tst[i+"_ENCODED"] = Final_tst[i].map(Name_encoded)
    Final_tst = Final_tst.drop(i+'_freq',axis=1)

drop_colm = []

for i in Final_tst.columns:
    if Final_tst[i].dtype == 'object':
        drop_colm.append(i)
        
Final_tst = Final_tst.drop(drop_colm,axis =1)
```

```python
def impute_nan_median(Final_tst,variable):
    median=Final_tst[variable].median()
    
    Final_tst[variable+"_NAN"]=np.where(Final_tst[variable].isnull(),1,0)
    Final_tst[variable+"_median"]=Final_tst[variable].fillna(median)
```

```python
p = Final_tst.isnull().sum(axis = 0)
p
```

```python
C_Names = ['HourlyDewPointTemperature','HourlyDryBulbTemperature','HourlyRelativeHumidity','HourlySeaLevelPressure','HourlyWindDirection','HourlyWindSpeed']
for i in C_Names:
    impute_nan_median(Final_tst,i)
```
